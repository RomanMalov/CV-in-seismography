```python
import torch


from torch.utils.data import Dataset
```

```python
import os
import numpy as np
class compressed(Dataset):
    def __init__(self, path_to_data, device, start = 0, fin = 1600):
        self.path_to_data =  path_to_data
        self.start = start
        self.fin = fin
        self.device = device
    def __len__(self):
        return self.fin - self.start

    def __getitem__(self, idx0):
        idx = self.start+idx0
        sample_path = os.path.join(self.path_to_data, 'sample_{idx}'.format(idx = idx))
        us = np.fromfile(sample_path+'/src/u.bin', dtype = 'float32').reshape((224, 30))
        ss = np.fromfile(sample_path+'/src/s.bin', dtype = 'float32')
        vhs = np.fromfile(sample_path+'/src/vh.bin', dtype = 'float32').reshape((30, 320))

        src = torch.tensor(us @ np.diag(ss) @ vhs, dtype = torch.float32)[None, :, :]
        
        ut = np.fromfile(sample_path+'/trg/u.bin', dtype = 'float32').reshape((224, 20))
        st = np.fromfile(sample_path+'/trg/s.bin', dtype = 'float32')
        vht = np.fromfile(sample_path+'/trg/vh.bin', dtype = 'float32').reshape((20, 320))

        
        trg = torch.tensor(ut @ np.diag(st) @ vht, dtype = torch.float32)

        return src, trg
        
    def split(self, split_ratio = [0.9 , 0.1]):
        assert sum(split_ratio) == 1
        idxes = [0]+[int(i*(self.fin - self.start)) for i in torch.cumsum(torch.tensor(split_ratio), dim = 0)]
        return [compressed(self.path_to_data, self.device, start = idxes[i], fin = idxes[i+1]) for i in range(len(split_ratio))]
```
